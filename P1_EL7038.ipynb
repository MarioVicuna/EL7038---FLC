```python
## Primera versión con skfuzzy, más adelante se hace todo a mano por relativa simplicidad

#!/usr/bin/env python2
# -*- coding: utf-8 -*-
import numpy as np
import skfuzzy.control as ctrl
import skfuzzy as fuzz
import matplotlib
import matplotlib.pyplot as plt

def de_a(A, B, literal = True):
    # Se considera que A y B son conjuntos difusos trapezoidales
    de_a = []
    if literal:
        de_a.append(A[0])
        de_a.append(A[1])
        de_a.append(B[2])
        de_a.append(B[3])
    else:
        for i in range(0,4):
            if i < 2:
                de_a.append(min(A[i],B[i]))
            else:
                de_a.append(max(A[i],B[i]))
    return de_a
```

```python
EP = ctrl.Antecedent(np.arange(-15, 15.05, 0.05), 'EP')
TP = ctrl.Antecedent(np.arange(-15, 15.05, 0.05), 'TP')
dH = ctrl.Consequent(np.arange(-15, 15.05, 0.05), 'dH')

names = ['Ng', 'Nm', 'Np', 'Ni', 'Ce', 'Pi', 'Pp', 'Pm', 'Pg']

Ng = (np.asarray([-1.0, -1.0, -0.7, -0.4]) * 15).tolist()
Nm = (np.asarray([-0.7, -0.5, -0.4, -0.2]) * 15).tolist()
Np = (np.asarray([-0.4, -0.3, -0.2, -0.1]) * 15).tolist()
Ni = (np.asarray([-0.2, -0.1, 0.0, 0.0]) * 15).tolist()
Ce = (np.asarray([-0.1, -0.0, 0.0, 0.1]) * 15).tolist()
Pi = (np.asarray([0.0, 0.0, 0.1, 0.2]) * 15).tolist()
Pp = (np.asarray([0.1, 0.2, 0.3, 0.4]) * 15).tolist()
Pm = (np.asarray([0.2, 0.4, 0.5, 0.7]) * 15).tolist()
Pg = (np.asarray([0.4, 0.7, 1.0, 1.0]) * 15).tolist()

NgUPp = de_a(Ng, Pp); names.append('NgUPp')
NgUNm = de_a(Ng, Nm); names.append('NgUNm')
NgUNp = de_a(Ng, Np); names.append('NgUNp')
NpUPi = de_a(Np, Pi); names.append('NpUPi')
NiUPi = de_a(Ni, Pi); names.append('NiUPi')
PmUPg = de_a(Pm, Pg); names.append('PmUPg')
NiUPp = de_a(Ni, Pp); names.append('NiUPp')
PpUPg = de_a(Pp, Pg); names.append('PpUPg')
NpUPg = de_a(Np, Pg); names.append('NpUPg')

EP['Ng'] = fuzz.trapmf(EP.universe, Ng)
EP['Nm'] = fuzz.trapmf(EP.universe, Nm)
EP['Np'] = fuzz.trapmf(EP.universe, Np)
EP['Ni'] = fuzz.trapmf(EP.universe, Ni)
EP['Ce'] = fuzz.trapmf(EP.universe, Ce)
EP['Pi'] = fuzz.trapmf(EP.universe, Pi)
EP['Pp'] = fuzz.trapmf(EP.universe, Pp)
EP['Pm'] = fuzz.trapmf(EP.universe, Pm)
EP['Pg'] = fuzz.trapmf(EP.universe, Pg)

EP.view()

EP['NgUPp'] = fuzz.trapmf(EP.universe, NgUPp)
EP['NgUNm'] = fuzz.trapmf(EP.universe, NgUNm)
EP['NgUNp'] = fuzz.trapmf(EP.universe, NgUNp)
EP['NpUPi'] = fuzz.trapmf(EP.universe, NpUPi)
EP['NiUPi'] = fuzz.trapmf(EP.universe, NiUPi)
EP['PmUPg'] = fuzz.trapmf(EP.universe, PmUPg)
EP['NiUPp'] = fuzz.trapmf(EP.universe, NiUPp)
EP['PpUPg'] = fuzz.trapmf(EP.universe, PpUPg)
EP['NpUPg'] = fuzz.trapmf(EP.universe, NpUPg)

EP.view()

TP['Ng'] = fuzz.trapmf(TP.universe, Ng)
TP['Nm'] = fuzz.trapmf(TP.universe, Nm)
TP['Np'] = fuzz.trapmf(TP.universe, Np)
TP['Ni'] = fuzz.trapmf(TP.universe, Ni)
TP['Ce'] = fuzz.trapmf(TP.universe, Ce)
TP['Pi'] = fuzz.trapmf(TP.universe, Pi)
TP['Pp'] = fuzz.trapmf(TP.universe, Pp)
TP['Pm'] = fuzz.trapmf(TP.universe, Pm)
TP['Pg'] = fuzz.trapmf(TP.universe, Pg)

TP.view()

TP['NgUPp'] = fuzz.trapmf(TP.universe, NgUPp)
TP['NgUNm'] = fuzz.trapmf(TP.universe, NgUNm)
TP['NgUNp'] = fuzz.trapmf(TP.universe, NgUNp)
TP['NpUPi'] = fuzz.trapmf(TP.universe, NpUPi)
TP['NiUPi'] = fuzz.trapmf(TP.universe, NiUPi)
TP['PmUPg'] = fuzz.trapmf(TP.universe, PmUPg)
TP['NiUPp'] = fuzz.trapmf(TP.universe, NiUPp)
TP['PpUPg'] = fuzz.trapmf(TP.universe, PpUPg)
TP['NpUPg'] = fuzz.trapmf(TP.universe, NpUPg)

TP.view()

dH['Ng'] = fuzz.trapmf(dH.universe, Ng)
dH['Nm'] = fuzz.trapmf(dH.universe, Nm)
dH['Np'] = fuzz.trapmf(dH.universe, Np)
dH['Ni'] = fuzz.trapmf(dH.universe, Ni)
dH['Ce'] = fuzz.trapmf(dH.universe, Ce)
dH['Pi'] = fuzz.trapmf(dH.universe, Pi)
dH['Pp'] = fuzz.trapmf(dH.universe, Pp)
dH['Pm'] = fuzz.trapmf(dH.universe, Pm)
dH['Pg'] = fuzz.trapmf(dH.universe, Pg)

dH.view()

dH['NgUPp'] = fuzz.trapmf(dH.universe, NgUPp)
dH['NgUNm'] = fuzz.trapmf(dH.universe, NgUNm)
dH['NgUNp'] = fuzz.trapmf(dH.universe, NgUNp)
dH['NpUPi'] = fuzz.trapmf(dH.universe, NpUPi)
dH['NiUPi'] = fuzz.trapmf(dH.universe, NiUPi)
dH['PmUPg'] = fuzz.trapmf(dH.universe, PmUPg)
dH['NiUPp'] = fuzz.trapmf(dH.universe, NiUPp)
dH['PpUPg'] = fuzz.trapmf(dH.universe, PpUPg)
dH['NpUPg'] = fuzz.trapmf(dH.universe, NpUPg)

dH.view()


rule1 = ctrl.Rule(antecedent = ((EP['Pg'] & TP['NpUPg']) |
                                (EP['PpUPg'] & TP['NgUNm'])),
                  consequent = dH['Ng'], label = 'rule Ng')

rule2 = ctrl.Rule(antecedent = ((EP['Pi'] & TP['PmUPg']) |
                                (EP['Pp'] & TP['NiUPp']) |
                                (EP['PmUPg'] & TP['PpUPg'])),
                  consequent = dH['Nm'], label = 'rule Nm')

rule3 = ctrl.Rule(antecedent = ((EP['Ni'] & TP['PmUPg']) |
                                (EP['Pi'] & TP['Pp'])),
                  consequent = dH['Np'], label = 'rule Np')

rule5 = ctrl.Rule(antecedent = ((EP['NiUPi'] & TP['Ce']) |
                                (EP['Ni'] & TP['Pp']) |
                                (EP['Pi'] & TP['Np'])),
                  consequent = dH['Ce'], label = 'rule Ce')

rule7 = ctrl.Rule(antecedent = ((EP['Pi'] & TP['NgUNm']) |
                                (EP['Ni'] & TP['Np'])),
                  consequent = dH['Pp'], label = 'rule Pp')

rule8 = ctrl.Rule(antecedent = ((EP['NgUNm'] & TP['NgUNp']) |
                                (EP['Np'] & TP['NpUPi']) |
                                (EP['Ni'] & TP['NgUNm'])),
                  consequent = dH['Pm'], label = 'rule Pm')

rule9 = ctrl.Rule(antecedent = ((EP['Ng'] & TP['NgUPp']) |
                                (EP['NgUNp'] & TP['PmUPg'])),
                  consequent = dH['Pg'], label = 'rule Pg')

rules = [rule1, rule2, rule3, rule5, rule7, rule8, rule9]

PO = 560
K = 0.6
initialConditions = [500, 550, 600]
defuzzMethods = ['mom', 'centroid', 'heights']

Pmom = []
Pcentroid = []
Pheights = []
```

```python
## MoM method for Pi = 500

dH.defuzzify_method = defuzzMethods[0]

system = ctrl.ControlSystem(rules)
sim = ctrl.ControlSystemSimulation(system)

Pi = initialConditions[0]
P = np.zeros(50)
P[0] = Pi
DH = np.zeros(50)


for i in range(1,50):
    
    EPt = (P[i - 1] - PO)
    if EPt > 15:
        EPt = 15
    elif EPt < -15:
        EPt = -15
        
    sim.input['EP'] = EPt
    
    if i == 1:
        sim.input['TP'] = 0
    else:
        TPt = P[i - 1] - P[i - 2]
        if TPt > 15:
            TPt = 15
        elif TPt < -15:
            TPt = -15
        sim.input['TP'] = TPt
    
    sim.compute()
    
    dHt = int(sim.output['dH'])
    if dHt > 15:
        dHt = 15
    elif dHt < -15:
        dHt = -15
    DH[i] = dHt
    dP = dHt * K
    P[i] = P[i - 1] + dP

Pmom.append(P)
plt.figure()
plt.plot(P,'-*')
plt.show()
```

```python
## Segunda versión sin skfuzzy. Se procede a redefinir todo para empezar.
import numpy
import matplotlib.pyplot as plt
import pandas as pd
import os
import xlsxwriter
import math

def de_a(A, B, literal = True):
    # Se considera que A y B son conjuntos difusos trapezoidales
    de_a = []
    if literal:
        de_a.append(A[0])
        de_a.append(A[1])
        de_a.append(B[2])
        de_a.append(B[3])
    else:
        for i in range(0,4):
            if i < 2:
                de_a.append(min(A[i],B[i]))
            else:
                de_a.append(max(A[i],B[i]))
    return de_a

def trapezoid_membership(x, Set):
    assert len(Set) == 4
    if -15.0 <= x <= Set[0]:
        if -15.0 == Set[0]:
            return 1
        return 0
    elif Set[0] <= x <= Set[1]:
        if Set[0] == Set[1]:
            m = 0
        else:
            m = 1/(Set[1] - Set[0])
        return m*x - m*Set[0]
    elif Set[1] <= x <= Set[2]:
        return 1
    elif Set[2] <= x <= Set[3]:
        if Set[0] == Set[1]:
            m = 0
        else:
            m = -1/(Set[3] - Set[2])
        return m*x - m*Set[2] + 1
    elif Set[3] <= x <= 15.0:
        return 0
    else:
        print("Out of bounds")
        return

def trapezoid_membership_vector(E, Set):
    output = []
    for i in range(0,len(E)):
        output.append(trapezoid_membership(E[i],Set))
    return output

def fuzzy_rule(E1, mf1, E2, mf2, domain, out_mf):
    c1 = trapezoid_membership_vector (E1, mf1)
    c2 = trapezoid_membership_vector(E2, mf2)
    c1_and_c2 = []
    for i in range(0,len(c1)):
        c1_and_c2.append(min(c1[i],c2[i]))
    out_function = trapezoid_membership_vector(domain, out_mf)
    output = []
    for i in range(0,len(c1_and_c2)):
        output.append(numpy.minimum(c1_and_c2[i],out_function))
    return numpy.asarray(output)

def fuzzy_rule_antecedent(E1, mf1, E2, mf2):
    c1 = trapezoid_membership_vector (E1, mf1)
    c2 = trapezoid_membership_vector(E2, mf2)
    c1_and_c2 = []
    for i in range(0,len(c1)):
        c1_and_c2.append(min(c1[i],c2[i]))
    return c1_and_c2

def aggregation(rule_map):
    output = []
    for row in range(0, len(rule_map[0])):
        aggregated_row = []
        for column in range(0, len(rule_map[0][0])):
            current_max = -math.inf
            for rule in range(0, len(rule_map)):
                if rule_map[rule][row][column] >= current_max:
                    current_max = rule_map[rule][row][column]
            aggregated_row.append(current_max)
        output.append(aggregated_row)
    return output

def COG(interval, function):
    output = []
    for row in range(0, len(function)):
        num = sum((function[row])*interval)
        den = sum(function[row])
        if num == 0 and den == 0:
            num = 0
        else:
            num = num/den
        output.append(num)
    return output

def MOM(interval, function):
    M = max(function[0])
    for i in range(0, len(function[0])):
        if function[0][i] != M:
            function[0][i] = 0
    return COG(interval, function)

def Heights(interval, function):
    return function
    
def defuzz(interval, function, method):
    return method(interval, function)
```
